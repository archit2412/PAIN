```python
import os
import numpy as np
import pandas as pd
import scipy.io
from scipy.signal import butter, filtfilt, iirnotch, resample, find_peaks
from scipy.stats import skew, kurtosis
import pywt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, LeaveOneGroupOut
# from sklearn.feature_selection import SelectFromModel # Not used as per design choice
from imblearn.over_sampling import SMOTE
import shap
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
import joblib  # Added for model saving

# Suppress specific warnings from sklearn/xgboost/etc. for cleaner output
warnings.filterwarnings("ignore", message="The 'use_label_encoder' parameter is deprecated and will be removed in a future release.")
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn.metrics._classification")
warnings.filterwarnings("ignore", message="The least populated class in y has only 1 sample(s) which is too few. The minimum number of samples for any class cannot be less than n_neighbors=5.")
warnings.filterwarnings("ignore", message="The number of training samples is less than the number of n_splits=5. Running 5 splits anyway.")


# --- Advanced Denoising Utilities ---
def bandpass_filter(data, fs, lowcut, highcut, order=4):
    """Applies a Butterworth bandpass filter."""
    if len(data) == 0:
        return np.array([])
    nyq = 0.5 * fs
    low, high = lowcut/nyq, highcut/nyq
    
    # Handle cases where low or high might exceed nyquist due to precision or too high target freq
    if high > 1.0: high = 0.99
    if low < 0.01: low = 0.01 # Ensure lowcut is not zero or too small relative to nyq
    if low >= high: return data # Avoid issues if filter range is invalid or frequencies cross
    
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data, axis=0)

def notch_filter(data, fs, freq=50.0, Q=30):
    """Applies a Butterworth notch filter."""
    if len(data) == 0:
        return np.array([])
    nyq = 0.5 * fs
    w0 = freq / nyq
    # Check if w0 is valid (0 < w0 < 1)
    if not (0 < w0 < 1):
        # print(f"Warning: Invalid notch frequency w0={w0}. Skipping notch filter.")
        return data
    b, a = iirnotch(w0, Q)
    return filtfilt(b, a, data, axis=0)

def wavelet_denoise(sig, wavelet='db4', level=2):
    """Applies wavelet denoising using soft thresholding."""
    if len(sig) < (2**(level+1)): # Ensure signal is long enough for wavelet decomposition at given level
        return sig # Cannot denoise if too short or level too high for signal length

    coeffs = pywt.wavedec(sig, wavelet, level=level)
    # Estimate sigma from the highest frequency detail coefficients
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745 # Using last detail coefficients
    uthresh = sigma * np.sqrt(2 * np.log(len(sig)))
    
    # Apply soft thresholding
    denoised_coeffs = [coeffs[0]] + [pywt.threshold(c, value=uthresh, mode='soft') for c in coeffs[1:]]
    
    # Reconstruct signal, ensuring output length matches input
    denoised_sig = pywt.waverec(denoised_coeffs, wavelet)
    return denoised_sig[:len(sig)] # Trim to original length in case of slight mismatch


# ICA for EMG artifact rejection (modified for standard denoising reconstruction)
def ica_denoise_emg(data):
    """
    Applies FastICA for denoising and reconstructs the signal.
    Assumes data is (n_samples, n_features/channels).
    If data has less than 2 features/channels, ICA is skipped.
    """
    try:
        from sklearn.decomposition import FastICA
        
        # ICA requires at least 2 features/channels for meaningful separation
        if data.shape[1] < 2:
            return data # Return original data if ICA not applicable
        
        # n_components_ica should not exceed min(n_samples, n_features)
        n_components_ica = data.shape[1]
        
        # Adjust n_components if there are fewer samples than channels
        if data.shape[0] < n_components_ica:
            n_components_ica = data.shape[0] 
        
        # If after checks, still not enough components for ICA (e.g., only 1 valid channel)
        if n_components_ica < 2: 
            return data
            
        # Increased tolerance to 0.001 as requested, while keeping max_iter at 10000
        ica = FastICA(n_components=n_components_ica, random_state=42, max_iter=10000, tol=0.001) 
        S_ = ica.fit_transform(data) # Estimated sources
        A_ = ica.mixing_ # Estimated mixing matrix (mixing sources back to observations)
        
        # Reconstruct the signal from the estimated sources and mixing matrix.
        reconstructed_data = np.dot(S_, A_.T) 
        
        return reconstructed_data
    except Exception as e:
        # print(f"Error during ICA denoising: {e}. Returning original data.")
        return data

# --- HRV, SCR, Poincaré, and Advanced Feature Extraction ---
def hrv_features(sig, fs):
    """Extracts Heart Rate Variability features from ECG signal."""
    sig = np.nan_to_num(sig)
    # Adjust peak detection for ECG signal characteristics (R-peaks are sharp and high)
    # Height is relative to the signal, distance ensures not detecting noise as multiple peaks
    peaks, _ = find_peaks(sig, height=np.std(sig)*0.5, distance=int(fs*0.4)) # min 0.4s between peaks
    
    rr = np.diff(peaks) / fs # RR intervals in seconds
    
    if len(rr) < 2: # Need at least 2 RR intervals for most features
        return [0] * 8 # Return zeros if not enough data
    
    hr = 60 / np.mean(rr) if np.mean(rr) > 0 else 0
    rmssd = np.sqrt(np.mean(np.square(np.diff(rr))))
    sdnn = np.std(rr)
    
    # Frequency domain (requires interpolation for regular sampling before FFT)
    if len(rr) > 2:
        t_rr = np.cumsum(rr)
        # Create a regular time axis for interpolation, common is 4 Hz
        t_interp = np.arange(t_rr[0], t_rr[-1], 1/4.0) 
        if len(t_interp) > 2: # Ensure enough points for meaningful interpolation
            rr_interp = np.interp(t_interp, t_rr, rr)
            
            # Ensure FFT input is at least 2 points
            if len(rr_interp) > 1:
                freqs = np.fft.rfftfreq(len(rr_interp), 1/4.0)
                psd = np.abs(np.fft.rfft(rr_interp))**2
                
                lf = np.sum(psd[(freqs >= 0.04) & (freqs < 0.15)])
                hf = np.sum(psd[(freqs >= 0.15) & (freqs < 0.4)])
                lf_hf = np.log1p(lf/(hf + 1e-9)) if hf > 0 else 0 # Add epsilon for stability
            else:
                lf = hf = lf_hf = 0
        else:
            lf = hf = lf_hf = 0
    else:
        lf = hf = lf_hf = 0

    # Poincaré Plot features (SD1, SD2)
    if len(rr) > 1:
        rr_n = rr[:-1]
        rr_n_plus_1 = rr[1:]
        # Standard deviation of points perpendicular to line of identity
        sd1 = np.sqrt(0.5 * np.nanvar(rr_n_plus_1 - rr_n)) 
        # Standard deviation of points along the line of identity
        sd2 = np.sqrt(0.5 * np.nanvar(rr_n_plus_1 + rr_n))
        # Handle NaN if variance is 0
        sd1 = np.nan_to_num(sd1)
        sd2 = np.nan_to_num(sd2)
    else:
        sd1 = 0.0
        sd2 = 0.0
    
    return [hr, rmssd, sdnn, lf, hf, lf_hf, sd1, sd2]

def scr_features(sig, fs):
    """Extracts Skin Conductance Response features from SCL signal."""
    sig = np.nan_to_num(sig)
    mean_scl = np.mean(sig)
    slope = (sig[-1] - sig[0]) / len(sig) if len(sig) > 1 else 0
    auc = np.trapz(sig) if len(sig) > 1 else 0 # Using np.trapz which is common for AUC

    # SCR detection: simple peak count above a threshold relative to signal mean/std
    # This can be improved with more sophisticated SCR decomposition algorithms if needed
    peak_height_threshold = np.mean(sig) + 2 * np.std(sig) # Peaks significantly above mean
    scrs, _ = find_peaks(sig, height=peak_height_threshold, distance=int(fs*1)) # Min 1s between SCRs

    scr_count = len(scrs)
    scr_amp = np.mean(sig[scrs]) if scr_count > 0 else 0

    return [mean_scl, scr_count, scr_amp, slope, auc]

def extract_ultra_features(data, fs, ch_names):
    """Extracts a comprehensive set of features from a multi-channel physiological segment."""
    feats = []
    
    # Identify EMG channels for ICA. Assuming first 3 are EMG based on typical setups.
    # Adjust this logic if your channel naming/ordering is different.
    emg_indices = [i for i, name in enumerate(ch_names) if 'emg' in name.lower() or 'corrugator' in name.lower() or 'zygomaticus' in name.lower()]
    
    if len(emg_indices) >= 2 and data.shape[0] > 1: # Need at least 2 EMG channels and enough samples
        emg_data_for_ica = data[:, emg_indices]
        denoised_emg_from_ica = ica_denoise_emg(emg_data_for_ica)
        # Store denoised EMG back into a temporary structure matching original indices
        denoised_emg_full_space = np.copy(data) # Start with a copy of original data
        denoised_emg_full_space[:, emg_indices] = denoised_emg_from_ica
    else:
        denoised_emg_full_space = np.copy(data) # If ICA not applied, use original data for processing

    processed_signals = {} # To store processed signals for cross-channel features

    for idx in range(data.shape[1]): # Iterate through all channels in the segment
        sig_original = data[:, idx] # Original raw signal
        channel_name = ch_names[idx] # Get the name of the current channel

        # Use ICA-denoised EMG signal if applicable, otherwise original
        if idx in emg_indices and denoised_emg_full_space.shape[1] > idx:
            sig = denoised_emg_full_space[:, idx]
        else:
            sig = sig_original

        # --- Re-introducing Filtering and Denoising ---
        # Apply filtering based on channel type (identified from name)
        if 'emg' in channel_name.lower() or 'corrugator' in channel_name.lower() or 'zygomaticus' in channel_name.lower():
            # Bandpass filter EMG to 20-450 Hz to remove motion artifacts and high-frequency noise
            sig = bandpass_filter(sig, fs, 20, 450)
            # Apply notch filter for power line interference (e.g., 50 Hz or 60 Hz depending on region)
            sig = notch_filter(sig, fs, 50) # Assuming 50 Hz, adjust if needed
        elif 'ecg' in channel_name.lower():
            # Bandpass filter ECG to 0.5-50 Hz for heart rate variability analysis
            sig = bandpass_filter(sig, fs, 0.5, 50)
            # Apply notch filter
            sig = notch_filter(sig, fs, 50) # Assuming 50 Hz, adjust if needed
        elif 'scl' in channel_name.lower() or 'gsr' in channel_name.lower():
            # Bandpass filter SCL (GSR) to 0.05-5 Hz to capture skin conductance responses while removing drift and high-freq noise
            sig = bandpass_filter(sig, fs, 0.05, 5)

        # Resample all signals to a common, lower sampling rate (e.g., 250 Hz) for consistency if original fs is higher
        target_fs = 250 # Hz
        if fs > target_fs and len(sig) > 1: # Ensure signal is long enough for resampling
            num_samples_resampled = int(len(sig) / (fs / target_fs))
            if num_samples_resampled > 1: # Ensure at least 2 samples after resampling
                sig = resample(sig, num_samples_resampled)
            else:
                sig = np.array([]) # If resampling leads to too few samples
        
        # Wavelet denoise as a final step for additional noise reduction
        sig = wavelet_denoise(sig)

        # Store the processed signal (even if empty) for later cross-channel features
        # Ensure it has the correct length before storing for cross-channel ops
        if len(sig) > 0:
            processed_signals[idx] = sig
        else:
            # If a signal becomes empty after processing, store an array of zeros matching original expected length
            processed_signals[idx] = np.zeros(data.shape[0]) 

        # Ensure signal is not empty after processing before feature extraction
        if len(sig) == 0:
            # Append zeros for all features if signal is empty, approximate count for safety
            num_common_feats = 7 + 5 + (3+1+1) # Time, Freq, Wavelet
            num_hrv_feats = 8
            num_scr_feats = 5

            feats.extend([0]*num_common_feats)
            if 'ecg' in channel_name.lower():
                feats.extend([0]*num_hrv_feats)
            elif 'scl' in channel_name.lower() or 'gsr' in channel_name.lower():
                feats.extend([0]*num_scr_feats)
            continue # Skip to next channel if current is empty

        # Time-domain features
        rms = np.sqrt(np.mean(sig**2))
        log_rms = np.log1p(rms)
        std = np.std(sig)
        wl = np.sum(np.abs(np.diff(sig))) if len(sig) > 1 else 0.0
        zcr = ((sig[:-1]*sig[1:])<0).sum() / len(sig) if len(sig) > 1 else 0.0
        feats.extend([rms, log_rms, std, wl, zcr, skew(sig) if len(sig) > 1 else 0.0, kurtosis(sig) if len(sig) > 1 else 0.0])
        
        # Frequency-domain features (using target_fs for FFT)
        if len(sig) > 1: # Ensure signal is long enough for FFT
            freqs = np.fft.rfftfreq(len(sig), 1/target_fs) 
            psd = np.abs(np.fft.rfft(sig))**2
            
            # Sum power in predefined frequency bands
            sum_0_5_4 = np.sum(psd[(freqs>=0.5)&(freqs<4)]) if np.any((freqs>=0.5)&(freqs<4)) else 0
            sum_4_8 = np.sum(psd[(freqs>=4)&(freqs<8)]) if np.any((freqs>=4)&(freqs<8)) else 0
            sum_8_13 = np.sum(psd[(freqs>=8)&(freqs<13)]) if np.any((freqs>=8)&(freqs<13)) else 0
            sum_13_30 = np.sum(psd[(freqs>=13)&(freqs<30)]) if np.any((freqs>=13)&(freqs<30)) else 0
            sum_30_45 = np.sum(psd[(freqs>=30)&(freqs<45)]) if np.any((freqs>=30)&(freqs<45)) else 0
            feats.extend([sum_0_5_4, sum_4_8, sum_8_13, sum_13_30, sum_30_45])
        else:
            feats.extend([0]*5) # Append zeros if FFT cannot be computed

        # Wavelet energy/entropy (using db4 wavelet, level 3)
        # 2**level must be less than signal length
        if len(sig) >= (2**(3+1)): # For level 3, we need at least 2^4 = 16 samples
            coeffs = pywt.wavedec(sig, 'db4', level=3)
            energies = [np.sum(np.square(c)) for c in coeffs]
            total_energy = np.sum(energies)
            # Handle division by zero for entropy if total_energy is zero
            entropy = -np.sum([(e/total_energy)*np.log(e/total_energy+1e-8) if e>0 and total_energy > 0 else 0 for e in energies])
            feats.extend(energies + [entropy])
        else:
            # Append zeros for the 4 energies + 1 entropy (total 5 features for level 3)
            feats.extend([0]*5) 

        # ECG HRV features
        if 'ecg' in channel_name.lower():
            feats.extend(hrv_features(sig, target_fs)) # Pass effective sampling rate
        # SCL SCR features
        if 'scl' in channel_name.lower() or 'gsr' in channel_name.lower():
            feats.extend(scr_features(sig, target_fs)) # Pass effective sampling rate
    
    # --- Cross-channel features (correlations, ratios) ---
    # Resample all processed signals to a common length (e.g., the original segment length)
    # This is critical if earlier resampling changed lengths, ensuring all arrays are alignable.
    original_segment_length = data.shape[0] # The original segment length from .mat
    processed_signal_array_aligned = []
    for idx in range(data.shape[1]):
        if idx in processed_signals:
            current_sig = processed_signals[idx]
            if len(current_sig) != original_segment_length and len(current_sig) > 0:
                # Resample the processed signal back to the original length if it changed
                current_sig = resample(current_sig, original_segment_length)
            elif len(current_sig) == 0:
                current_sig = np.zeros(original_segment_length) # Use zeros if signal became empty
            processed_signal_array_aligned.append(current_sig)
        else:
            processed_signal_array_aligned.append(np.zeros(original_segment_length)) # Should not happen with current logic

    if len(processed_signal_array_aligned) == data.shape[1] and data.shape[1] > 1:
        processed_data_stacked = np.array(processed_signal_array_aligned).T # Stack to (samples, channels)
        
        for i in range(processed_data_stacked.shape[1]):
            for j in range(i+1, processed_data_stacked.shape[1]):
                # Correlation
                # Ensure signals have variance for meaningful correlation
                if processed_data_stacked[:,i].std() > 0 and processed_data_stacked[:,j].std() > 0:
                    feats.append(np.corrcoef(processed_data_stacked[:,i], processed_data_stacked[:,j])[0,1])
                else:
                    feats.append(0) # Correlation is undefined if std dev is zero
                
                # Ratio of means
                mean_i = np.mean(processed_data_stacked[:,i])
                mean_j = np.mean(processed_data_stacked[:,j])
                feats.append(mean_i / (mean_j + 1e-8)) # Add small epsilon to avoid division by zero
    else:
        # Append zeros for cross-channel features if not enough channels or processed signals
        num_cross_channel_pairs = (data.shape[1] * (data.shape[1] - 1)) // 2
        feats.extend([0] * (num_cross_channel_pairs * 2)) # 2 features per pair (corr, ratio)


    # Delta features (change from start to end)
    # Computed on *processed* signals after all filtering and resampling
    for idx in range(data.shape[1]):
        if idx in processed_signals and len(processed_signals[idx]) > 1:
            feats.append(processed_signals[idx][-1] - processed_signals[idx][0])
        else:
            feats.append(0) # Append zero if signal not processed or too short

    return np.nan_to_num(feats) # Convert any NaN or Inf to numbers


# --- CORAL Domain Adaptation ---
def coral(source, target):
    """
    Splits a dictionary of parameters into sub-dictionaries based on prefixes.
    For example, {'xgb__param1': val1, 'lgbm__param2': val2, 'param3': val3}
    with prefix 'xgb__' would return {'param1': val1}.
    """
    if source.shape[0] < source.shape[1] or target.shape[0] < target.shape[1] or \
       np.linalg.matrix_rank(np.cov(source, rowvar=False)) < source.shape[1] or \
       np.linalg.matrix_rank(np.cov(target, rowvar=False)) < target.shape[1]:
        # print("Warning: Not enough samples or full rank for covariance calculation in CORAL. Skipping.")
        return source # Return source as is if not enough samples or singular covariance
    
    cov_s = np.cov(source, rowvar=False) + np.eye(source.shape[1]) * 1e-6 # Add small epsilon for stability
    cov_t = np.cov(target, rowvar=False) + np.eye(target.shape[1]) * 1e-6
    
    try:
        # Cholesky decomposition of covariance matrices
        chol_s_inv = np.linalg.cholesky(cov_s)
        chol_t = np.linalg.cholesky(cov_t)
        
        # A_coral = (L_target @ L_source_inv)
        A_coral = np.dot(chol_t, np.linalg.inv(chol_s_inv))
        
        # Apply CORAL transformation
        # Data is centered first, transformed, then target mean is added back
        return np.dot(source - np.mean(source, axis=0), A_coral) + np.mean(target, axis=0)
    except np.linalg.LinAlgError as e:
        # print(f"Warning: Linear algebra error in CORAL ({e}). Skipping CORAL for this fold.")
        return source # Fallback to original source data on error
    except ValueError as e:
        # print(f"Warning: ValueError in CORAL ({e}). Skipping CORAL for this fold.")
        return source # Fallback for other potential errors (e.g., non-positive definite covariance)

# Helper function to extract parameters for a specific estimator from the best_params_
def get_estimator_params(best_params, prefix):
    """
    Extracts parameters for a specific estimator from the best_params_ dictionary
    returned by GridSearchCV/RandomizedSearchCV.
    """
    return {key.replace(prefix, ''): value for key, value in best_params.items() if key.startswith(prefix)}

# --- Data Loader ---
def collect_segments_ultra(root_dir):
    """
    Collects physiological segments, labels, and subject IDs from .mat files
    organized by pain label and subject ID. Applies feature extraction.
    """
    segments, labels, subjects, pain_map = [], [], [], {}
    
    # Discover subject IDs dynamically by iterating through existing folders
    all_subject_ids = set()
    try:
        for pain_label in os.listdir(root_dir):
            pain_path = os.path.join(root_dir, pain_label)
            if os.path.isdir(pain_path):
                for subj in os.listdir(pain_path):
                    if os.path.isdir(os.path.join(pain_path, subj)):
                        all_subject_ids.add(subj)
        all_subject_ids = sorted(list(all_subject_ids))
    except FileNotFoundError:
        print(f"Error: Root directory '{root_dir}' not found. Please check the path.")
        return np.array([]), np.array([]), np.array([]), {}
    except Exception as e:
        print(f"An error occurred during subject discovery: {e}")
        return np.array([]), np.array([]), np.array([]), {}


    print(f"Discovered {len(all_subject_ids)} subjects: {', '.join(all_subject_ids)}")
    if not all_subject_ids:
        print("No subjects found. Please check your data directory structure.")
        return np.array([]), np.array([]), np.array([]), {}


    # Collect pain labels from the root directory
    available_pain_labels = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
    
    print("\nStarting feature extraction from .mat files...")
    # Outer loop for pain labels (e.g., 'low_pain', 'med_pain')
    for pain_label in tqdm(available_pain_labels, desc="Processing Pain Labels"):
        pain_path = os.path.join(root_dir, pain_label)
        
        # Inner loop for subjects to ensure LOSO compatibility later
        for subj in all_subject_ids: # Iterate over ALL discovered subjects
            subj_path = os.path.join(pain_path, subj)
            
            if not os.path.isdir(subj_path): 
                # print(f"Skipping {subj_path}: Directory not found for this pain label.")
                continue # Skip if subject folder doesn't exist for this pain level

            for segfile in sorted(os.listdir(subj_path)):
                if not segfile.endswith('.mat'): continue
                matf = os.path.join(subj_path, segfile)
                try:
                    mat = scipy.io.loadmat(matf)
                    
                    # Ensure 'data' field exists and is not empty
                    if 'data' not in mat or mat['data'].size == 0:
                        # print(f"Skipping {matf}: 'data' field not found or is empty.")
                        continue
                    data = mat['data']

                    # Extract sampling frequency (fs) and channel labels
                    fs = int(mat['fs'].ravel()[0]) if 'fs' in mat and mat['fs'].size > 0 and mat['fs'].ndim > 0 else 1000 # Default to 1000 Hz if not found or malformed
                    ch_names = [str(s[0]) if isinstance(s, np.ndarray) else str(s) for s in mat['labels'].ravel()] if 'labels' in mat and mat['labels'].size > 0 and mat['labels'].ndim > 0 else [f'ch{i}' for i in range(data.shape[1])]
                    
                    # Ensure data is 2D (samples x channels)
                    if data.ndim == 1:
                        data = data[:, np.newaxis]
                    elif data.ndim > 2:
                        data = data.squeeze() # Remove singleton dimensions if any

                    # Ensure data has enough samples for processing (e.g., at least 1 second for 1000Hz)
                    if data.shape[0] < fs: # For a 1-second segment minimum
                        # print(f"Skipping {matf}: Not enough samples for processing (length: {data.shape[0]} at {fs}Hz).")
                        continue

                    feats = extract_ultra_features(data, fs, ch_names)
                    
                    # Only add if features were actually extracted and are not all zeros (indicating an issue)
                    if len(feats) > 0 and not np.all(np.array(feats) == 0): 
                        segments.append(feats)
                        labels.append(pain_label)
                        subjects.append(subj)
                        # Map pain labels to integers (e.g., 'low_pain': 0, 'med_pain': 1)
                        if pain_label not in pain_map:
                            pain_map[pain_label] = len(pain_map)
                    # else:
                        # print(f"Skipping {matf}: No valid features extracted or all features are zero after processing.")
                except Exception as e:
                    print(f"Error processing {matf}: {e}")
    
    if not segments:
        print("No segments loaded. Please check your data directory, file structure, and .mat file contents.")
        return np.array([]), np.array([]), np.array([]), {}

    return np.array(segments), np.array(labels), np.array(subjects), pain_map

# --- Main Pipeline Execution ---
if __name__ == "__main__":
    print("Starting the complete Pain Prediction Pipeline...")

    # --- Configuration ---
    # !!! IMPORTANT: Set this to the root directory of your .mat files !!!
    # Example structure: D:\bio_s\low_pain\S001\segment1.mat, D:\bio_s\med_pain\S001\segment2.mat
    root_dir = r"D:\biodire" 

    # --- 1. Load Data and Extract Features ---
    X, y, subjects, pain_map = collect_segments_ultra(root_dir)

    if len(X) == 0:
        raise RuntimeError("No valid segments loaded. Please check your data directory and preprocessing functions.")
    
    print(f"Total segments loaded: {len(X)}, Features extracted per segment: {X.shape[1]}")
    
    # Map string labels to integer encodings
    label_names = sorted(pain_map, key=lambda k: pain_map[k]) # Ensures order is consistent
    y_enc = np.array([pain_map[lab] for lab in y])

    # --- 2. Train/Test Split (Standard Evaluation) ---
    print("\n--- Performing Standard Train/Test Split Evaluation ---")
    # Stratify by y to maintain class proportions in splits
    # Adjust test_size based on your dataset size and desired split
    X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size=0.3, stratify=y_enc, random_state=42)

    # Per-split normalization (fit on train, transform both)
    scaler_split = StandardScaler()
    X_train_scaled = scaler_split.fit_transform(X_train)
    X_test_scaled = scaler_split.transform(X_test)

    # Class Balancing (SMOTE) only on training data
    sm = SMOTE(random_state=42)
    unique_labels_train, counts_train = np.unique(y_train, return_counts=True)
    min_samples_in_class_train = np.min(counts_train) if len(counts_train) > 0 else 0

    if min_samples_in_class_train < sm.k_neighbors + 1:
        if min_samples_in_class_train > 1:
            sm_adjusted = SMOTE(random_state=42, k_neighbors=min_samples_in_class_train - 1)
            X_train_balanced, y_train_balanced = sm_adjusted.fit_resample(X_train_scaled, y_train)
            print(f"Adjusted SMOTE k_neighbors to {min_samples_in_class_train - 1} for train/test split.")
        else:
            print("Cannot apply SMOTE for train/test split, smallest class has 1 or 0 samples. Skipping SMOTE.")
            X_train_balanced, y_train_balanced = X_train_scaled, y_train # Fallback
    else:
        X_train_balanced, y_train_balanced = sm.fit_resample(X_train_scaled, y_train)
    print(f"Standard Split - Original Train Shape: {X_train.shape}, Balanced Train Shape: {X_train_balanced.shape}")

    # Domain Adaptation (CORAL) - align training data to test data distribution
    if X_train_balanced.shape[0] >= X_train_balanced.shape[1] and X_test_scaled.shape[0] >= X_test_scaled.shape[1]: 
        X_train_adapted = coral(X_train_balanced, X_test_scaled)
        print("CORAL applied for standard train/test split.")
    else:
        X_train_adapted = X_train_balanced # Fallback
        print("Not enough samples for CORAL in standard train/test split. Skipping CORAL.")


    # --- Stacking Ensemble Definition ---
    # Define base learners
    xgb_base = XGBClassifier(n_estimators=200, max_depth=6, eval_metric='logloss', use_label_encoder=False, random_state=42, n_jobs=20) # n_jobs=20 for max CPU usage
    lgbm_base = LGBMClassifier(n_estimators=200, max_depth=6, class_weight='balanced', random_state=42, n_jobs=20) # n_jobs=20 for max CPU usage
    svm_base = SVC(kernel='rbf', probability=True, class_weight='balanced', random_state=42)

    base_learners = [('xgb', xgb_base), ('lgbm', lgbm_base), ('svm', svm_base)]
    meta_learner = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=20) # n_jobs=20 for max CPU usage

    # Define the StackingClassifier for tuning
    stack_for_tuning = StackingClassifier(estimators=base_learners, final_estimator=meta_learner, n_jobs=20, passthrough=True) # n_jobs=20 for max CPU usage


    # --- Hyperparameter Tuning with RandomizedSearchCV ---
    print("\n--- Starting Hyperparameter Tuning for Stacking Classifier ---")

    param_dist = {
        'xgb__n_estimators': [100, 200, 300],
        'xgb__max_depth': [3, 6, 9],
        'xgb__learning_rate': [0.01, 0.05, 0.1],
        'lgbm__n_estimators': [100, 200, 300],
        'lgbm__max_depth': [3, 6, 9],
        'lgbm__learning_rate': [0.01, 0.05, 0.1],
        'svm__C': [0.1, 1, 10],
        'svm__gamma': ['scale', 'auto'],
        'final_estimator__n_estimators': [50, 100, 200],
        'final_estimator__max_depth': [None, 5, 10],
    }

    random_search = RandomizedSearchCV(
        estimator=stack_for_tuning,
        param_distributions=param_dist,
        n_iter=50, # Number of parameter settings that are sampled. Adjust based on compute.
        cv=5,      # 5-fold cross-validation during tuning.
        scoring='accuracy',
        random_state=42,
        n_jobs=20, # Use all 20 logical processors
        verbose=1
    )

    random_search.fit(X_train_adapted, y_train_balanced) # Fit on adapted training data

    best_stack_model = random_search.best_estimator_
    print(f"\nBest Stacking Classifier Parameters: {random_search.best_params_}")
    print(f"Best Cross-Validation Accuracy during Tuning: {random_search.best_score_:.3f}")

    # Use the best model found by RandomizedSearchCV for final evaluation on the standard test set
    final_model_for_test = best_stack_model 

    # --- Save the Trained Model ---
    model_save_path = 'final_pain_prediction_stacking_model.joblib'
    joblib.dump(final_model_for_test, model_save_path)
    print(f"\nFinal trained StackingClassifier model saved to {model_save_path}")

    # --- Evaluation for Standard Split ---
    y_pred_standard = final_model_for_test.predict(X_test_scaled) # Predict on scaled test data
    acc_standard = accuracy_score(y_test, y_pred_standard)
    print(f"\nStandard Train/Test Split Accuracy: {acc_standard:.3f}")
    print("Classification Report (Standard Split):")
    print(classification_report(y_test, y_pred_standard, target_names=label_names))
    print("Confusion Matrix (Standard Split):")
    print(confusion_matrix(y_test, y_pred_standard)
)
    # ROC Curves for Standard Split
    try:
        y_proba_standard = final_model_for_test.predict_proba(X_test_scaled)
        plt.figure(figsize=(8, 6))
        for i, name in enumerate(label_names):
            # Only plot if the class exists in y_test
            if (y_test == i).astype(int).sum() > 0:
                fpr, tpr, _ = roc_curve((y_test==i).astype(int), y_proba_standard[:,i])
                plt.plot(fpr, tpr, label=f"ROC {name} (AUC={roc_auc_score((y_test==i).astype(int), y_proba_standard[:,i]):.2f})")
        plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
        plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate'); plt.title('ROC Curves (Standard Split)'); plt.legend(); plt.show()
    except Exception as e:
        print(f"ROC plot error for standard split: {e}. Check if predict_proba is supported or if test set has only one class.")

    # SHAP Feature Importance (using the best model from tuning)
    print("\nGenerating SHAP plots for base learners (may take a moment)...")
    feature_names_list = [f'feature_{i}' for i in range(X.shape[1])] # Generic feature names

    try:
        # Check if XGBoost base learner is present and fitted
        if 'xgb' in final_model_for_test.named_estimators_ and hasattr(final_model_for_test.named_estimators_['xgb'], 'get_booster'):
            explainer_xgb = shap.TreeExplainer(final_model_for_test.named_estimators_['xgb'])
            shap_values_xgb = explainer_xgb.shap_values(X_test_scaled)
            # shap_values can be a list if multi-class, take the first class's values for summary plot
            if isinstance(shap_values_xgb, list):
                shap.summary_plot(shap_values_xgb[0], X_test_scaled, feature_names=feature_names_list, show=False)
            else:
                shap.summary_plot(shap_values_xgb, X_test_scaled, feature_names=feature_names_list, show=False)
            plt.title('SHAP Feature Importance (XGBoost Base Learner - Standard Split)')
            plt.show()
        else:
            print("XGBoost base learner not found or not suitable for SHAP in stack.")
    except Exception as e:
        print(f"SHAP XGB error (standard split): {e}")

    try:
        # Check if LightGBM base learner is present and fitted
        if 'lgbm' in final_model_for_test.named_estimators_ and hasattr(final_model_for_test.named_estimators_['lgbm'], 'booster_'):
            explainer_lgbm = shap.TreeExplainer(final_model_for_test.named_estimators_['lgbm'])
            shap_values_lgbm = explainer_lgbm.shap_values(X_test_scaled)
            if isinstance(shap_values_lgbm, list):
                shap.summary_plot(shap_values_lgbm[0], X_test_scaled, feature_names=feature_names_list, show=False)
            else:
                shap.summary_plot(shap_values_lgbm, X_test_scaled, feature_names=feature_names_list, show=False)
            plt.title('SHAP Feature Importance (LightGBM Base Learner - Standard Split)')
            plt.show()
        else:
            print("LightGBM base learner not found or not suitable for SHAP in stack.")
    except Exception as e:
        print(f"SHAP LGBM error (standard split): {e}")


    # --- 3. Robust LOSO CV with Per-Fold Normalization, Balancing, CORAL ---
    print("\n--- Starting Robust LOSO Cross-Validation ---")
    
    # Initialize a new StackingClassifier for each fold using the best parameters found
    # This ensures fresh models for each fold.
    
    # Extract best parameters for each estimator
    xgb_best_params = get_estimator_params(random_search.best_params_, 'xgb__')
    lgbm_best_params = get_estimator_params(random_search.best_params_, 'lgbm__')
    svm_best_params = get_estimator_params(random_search.best_params_, 'svm__')
    final_estimator_best_params = get_estimator_params(random_search.best_params_, 'final_estimator__')

    base_learners_loso = [
        ('xgb', XGBClassifier(**xgb_best_params, eval_metric='logloss', use_label_encoder=False, random_state=42, n_jobs=20)),
        ('lgbm', LGBMClassifier(**lgbm_best_params, class_weight='balanced', random_state=42, n_jobs=20)),
        ('svm', SVC(**svm_best_params, probability=True, class_weight='balanced', random_state=42))
    ]
    meta_learner_loso = RandomForestClassifier(**final_estimator_best_params, random_state=42, n_jobs=20)

    loso_accuracies = []
    loso_all_y_true = []
    loso_all_y_pred = []

    loso = LeaveOneGroupOut()
    
    # Use the subject IDs as groups for LOSO CV
    # Ensure 'subjects' array aligns with X and y_enc (same order and length)

    # Corrected iteration: directly iterate over loso.split and then wrap with tqdm
    for fold_idx, (train_idx, test_idx) in enumerate(tqdm(loso.split(X, y_enc, groups=subjects), total=loso.get_n_splits(groups=subjects), desc="LOSO CV Progress")):
        X_train_fold, X_test_fold = X[train_idx], X[test_idx]
        y_train_fold, y_test_fold = y_enc[train_idx], y_enc[test_idx]

        # ... (rest of your LOSO loop code) ...

        # Skip fold if test set is empty (can happen if a subject has no valid segments)
        if len(y_test_fold) == 0:
            print(f"Warning: Test set for subject {subjects[test_idx][0]} is empty. Skipping fold.")
            continue
        # Skip fold if training set is too small to contain all classes for stratification/SMOTE
        unique_classes_train_fold, counts_train_fold = np.unique(y_train_fold, return_counts=True)
        if len(unique_classes_train_fold) < len(label_names) or np.min(counts_train_fold) < 2:
            print(f"Warning: Training set for subject {subjects[test_idx][0]} has too few samples per class ({np.min(counts_train_fold)}). Skipping SMOTE and potentially this fold for robustness.")
            # If SMOTE cannot be applied, ensure model can still train
            sm_fold_instance = None # Indicate SMOTE cannot be used for this fold
            X_train_balanced_fold, y_train_balanced_fold = X_train_fold, y_train_fold
        else:
            # Per-fold normalization
            scaler_fold = StandardScaler()
            X_train_scaled_fold = scaler_fold.fit_transform(X_train_fold)
            X_test_scaled_fold = scaler_fold.transform(X_test_fold)

            # Balancing (apply SMOTE only on training data)
            sm_k_neighbors = min(sm.k_neighbors, np.min(counts_train_fold) - 1) if np.min(counts_train_fold) > 1 else 1
            sm_fold_instance = SMOTE(random_state=42, k_neighbors=sm_k_neighbors)
            X_train_balanced_fold, y_train_balanced_fold = sm_fold_instance.fit_resample(X_train_scaled_fold, y_train_fold)


        # CORAL (align training data to test data distribution for the current fold)
        if X_train_balanced_fold.shape[0] >= X_train_balanced_fold.shape[1] and X_test_scaled_fold.shape[0] >= X_test_scaled_fold.shape[1]:
            X_train_adapted_fold = coral(X_train_balanced_fold, X_test_scaled_fold)
        else:
            X_train_adapted_fold = X_train_balanced_fold # Fallback
            # print(f"Warning: Not enough samples for CORAL in LOSO fold (subj: {subjects[test_idx][0]}). Skipping.")


        # Fit model
        # Create a new StackingClassifier instance for each fold
        stack_fold = StackingClassifier(estimators=base_learners_loso, final_estimator=meta_learner_loso, n_jobs=-1, passthrough=True)
        try:
            stack_fold.fit(X_train_adapted_fold, y_train_balanced_fold)
            y_pred_fold = stack_fold.predict(X_test_scaled_fold)
            
            loso_accuracies.append(accuracy_score(y_test_fold, y_pred_fold))
            loso_all_y_true.extend(y_test_fold)
            loso_all_y_pred.extend(y_pred_fold)
            
            progress_bar_loso.set_description(f"LOSO CV Progress (Fold {fold_idx+1}/{loso.get_n_splits(groups=subjects)}, Acc: {loso_accuracies[-1]:.3f})")
        except Exception as e:
            print(f"Error during model fitting/prediction for subject {subjects[test_idx][0]}: {e}. Skipping fold.")


    if loso_accuracies: # Check if any folds were successfully processed
        loso_mean_acc = np.mean(loso_accuracies)
        print(f"\nLOSO Mean Accuracy: {loso_mean_acc:.3f}")
        print("\nLOSO Classification Report (Aggregated):")
        print(classification_report(loso_all_y_true, loso_all_y_pred, target_names=label_names))
        print("LOSO Confusion Matrix (Aggregated):")
        print(confusion_matrix(loso_all_y_true, loso_all_y_pred))
    else:
        print("\nNo successful LOSO folds completed. Please review data and processing steps.")

    print("\n--- Pipeline Summary ---")
    print("This script implements a comprehensive pain prediction pipeline with:")
    print("1. **Direct Data Loading:** Configured to read your `.mat` files from `D:\bio_s` (configurable `root_dir`).")
    print("2. **Robust Preprocessing:** Includes bandpass filtering, notch filtering, wavelet denoising, and improved ICA for EMG (with adjusted tolerance).")
    print("3. **Extensive Feature Engineering:** Extracts time-domain, frequency-domain, wavelet-domain, HRV, SCR, cross-channel, and delta features.")
    print("4. **Domain Adaptation (CORAL):** Applied within splits/folds to enhance generalization.")
    print("5. **Class Imbalance Handling (SMOTE):** Dynamically adjusted and applied within cross-validation.")
    print("6. **Optimized Ensemble Model:** A `StackingClassifier` with tuned `XGBoost`, `LightGBM`, `SVC` base learners, and `RandomForestClassifier` meta-learner.")
    print("7. **Rigorous Evaluation:** Both standard train/test split and **Leave-One-Subject-Out (LOSO) Cross-Validation** are performed, with all preprocessing and balancing steps correctly applied per-fold.")
    print("8. **Interpretability:** Provides ROC curves and SHAP explanations for base learners.")
    print("\nThis pipeline is designed for high performance, generalization, and transparency in physiological signal-based pain prediction.")
```
